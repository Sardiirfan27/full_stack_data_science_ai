```python
def count_misval(df=None, show='both'):
    """
    Count missing values in a DataFrame and display based on the specified condition.

    Parameters:
        df (DataFrame): The DataFrame to analyze.
        show (str): Specifies which DataFrame to return. Possible values are 'both', 'misval', or 'non_misval'.
                    Default is 'both'.
                    
                    - 'misval': Returns a DataFrame containing columns with missing values.
                    - 'non_misval': Returns a DataFrame containing columns with no missing values.
                    - 'both': Returns a DataFrame containing all columns, regardless of missing values.

    Returns:
        DataFrame or str: Depending on the 'show' parameter, returns a DataFrame containing missing values 
                          information or a string indicating an invalid parameter value.
                          
    Example:
        Assuming df is your DataFrame:
        
        >>> result = count_misval(df, show='both')
        >>> print(result)
    """
    if df is None:
        return "DataFrame is not provided."
    
    # Calculate the total missing value in each column
    missing_values = df.isnull().sum()

    # Calculate the percentage of missing values in each column
    percentage_missing_values = (missing_values / len(df)) * 100

    # Create a new dataframe to store the total and percentage of missing values
    missing_df = pd.DataFrame(
        {'total missing values': missing_values, 
         'percentage': percentage_missing_values.round(2)})
    
    if show == 'misval': #
        return missing_df[missing_df['total missing values'] > 0]
    elif show == 'non_misval':
        return missing_df[missing_df['total missing values'] == 0]
    elif show == 'both':
        return missing_df
    else:
        return "Invalid value for 'show' parameter. Use 'both', 'misval', or 'non_misval'"
```

```python
def stacked_bar_chart(ax, ct_index, ct_columns, xlabel, ylabel, colors, color_text='black', fontsize=12):
    """
    Create a stacked bar chart with counts and percentage annotations.

    Parameters:
    -----------
    ax : matplotlib.axes.Axes
        The axes to plot on.
    ct_index : pd.Series
        Categories for the x-axis.
    ct_columns : pd.Series
        Categories for the stacked bars.
    xlabel : str
        Label for x-axis.
    ylabel : str
        Label for y-axis.
    colors : list
        List of colors for the stacked bars.
    color_text : str, optional
        Text color for annotations (default is 'black').
    fontsize : int, optional
        Font size for annotations (default is 12).
    """

    # Crosstab for counts
    ct_data_count = pd.crosstab(index=ct_index, columns=ct_columns)

    # Crosstab for proportions
    ct_data_prob = pd.crosstab(index=ct_index, columns=ct_columns, normalize='index')

    # Plot stacked bar chart
    ct_data_prob.plot(kind='bar', stacked=True, color=colors, ax=ax)

    # Labels
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xticklabels(ct_data_prob.index.values, rotation=0)
    ax.grid(axis='x', linestyle='', linewidth=0)

    # Add counts and percentages
    for n, category in enumerate(ct_data_prob.index.values):
        y_offset = 0
        for (proportion, count) in zip(ct_data_prob.loc[category], ct_data_count.loc[category]):
            if proportion > 0:  # avoid placing text on 0% bars
                ax.text(
                    x=n,
                    y=y_offset + proportion / 2,
                    s=f'{count}\n({proportion * 100:.1f}%)',
                    ha='center',
                    color=color_text,
                    fontsize=fontsize,
                    fontweight="bold"
                )
            y_offset += proportion
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df['MultipleLines'], df['Churn'])
contingency_table
```

```python
def chi_square_test(
    df: pd.DataFrame,
    target: str,
    features: List[str],
    alpha: float = 0.05
) -> pd.DataFrame:
    """
    Perform Chi-square tests for categorical features against a target variable.
    
    Parameters:
        df (pd.DataFrame): The input dataframe containing the data.
        target (str): The target column name.
        features (List[str]): List of feature column names to test.
        alpha (float): Significance level (default is 0.05).
    
    Returns:
        pd.DataFrame: A DataFrame containing p-values and test results.
    """
    results = []
    for feature in features:
        # Create contingency table
        contingency_table = pd.crosstab(df[feature], df[target])
        
        # Perform chi-square test
        chi2, p_value, _, _ = chi2_contingency(contingency_table)
        
        # Calculate Cram√©r's V
        n = contingency_table.sum().sum()
        min_dim = min(contingency_table.shape) - 1
        cramers_v = np.sqrt(chi2 / (n * min_dim))
        
        # Interpret relationship strength
        if cramers_v <= 0.10:
            strength = "Very Weak"
        elif cramers_v <= 0.30:
            strength = "Weak"
        elif cramers_v <= 0.50:
            strength = "Moderate"
        else:
            strength = "Strong"
        
        # Interpret hypothesis testing
        reject_h0 = "yes" if p_value <= alpha else "no"
        
        # Append results
        results.append({
            'feature': feature,
            'target': target,
            'p-value': round(p_value, 4),
            'reject_h0': reject_h0,
            'cramers_v': round(cramers_v, 4),
            'relationship_strength': strength
        })
    
    # Create DataFrame
    output_df = pd.DataFrame(results)
    return output_df

target = 'Churn'
features = ['MultipleLines'] 
output_table = chi_square_test(df, target, features)
display(output_table)
```

  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
    From the results of our Exploratory Data Analysis, we identified several features that have a significant impact on churn, such as Internet Service, Tenure, and others. However, there are also features that appear to be correlated but may actually be influenced by other variables. In this section, we will re-examine all the features and apply various feature engineering and preprocessing techniques before building the machine learning model.</p>
</div>


```python
df_new= df.copy()
```

```python
# encode target
target='Churn'
df_new[target]=df['Churn'].map({'No':0, 'Yes':1})
```

```python
def spearman_corr_df(df, numerical_features, target_col):
    results = []
    for feature in numerical_features:
        corr, _ = spearmanr(df[feature], df[target_col])
        results.append({
            'Features': feature,
            'Spearman_Correlation': corr,
            'Abs_Correlation': abs(corr)
        })
    
    return (pd.DataFrame(results)
            .sort_values(by='Abs_Correlation', ascending=False)
            .reset_index(drop=True)
           )

num_features_corr= spearman_corr_df(df=df_new,numerical_features=numerical_features, target_col=target)
num_features_corr
```

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
    For numerical features, we conclude that tenure is the most correlated feature with churn. However, during the EDA, we also found insights showing that other features such as Total Charges and Monthly Charges are correlated with tenure. To address this, we can also perform a multicollinearity check using the Variance Inflation Factor (VIF).</p>
</div>


```python
def calc_vif(X):
    vif = pd.DataFrame()
    vif["Feature"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif["Level"] = vif["VIF"].apply(
        lambda v: "Low" if v < 5 
        else "Medium" if v < 10 
        else "High"
    )
    
    return (vif
            .sort_values(by='VIF', ascending=False)
            .reset_index(drop=True)
           )

calc_vif(df_new[numerical_features])
```

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
   From the VIF results, we can observe that Total Charges and Tenure exhibit a moderate level of multicollinearity, while Monthly Charges shows a relatively low level of multicollinearity. A moderate level indicates that there is some correlation among these features, which should be taken into account when building the machine learning model. Considering this factor allows us to develop a more efficient and robust model. In the modeling stage, we will compare the evaluation results between using all features and using a reduced set of features, either by removing one of the highly collinear features or by applying dimensionality reduction with PCA.</p>
</div>


<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
   From a mathematical perspective, Total Charges can essentially be derived from the multiplication of tenure and Monthly Charges, adjusted by potential discounts. Therefore, we can remove Total Charges, as its information is already represented by tenure and Monthly Charges.</p>
</div>


```python
numerical_features_fe1 = [f for f in numerical_features if f != 'TotalCharges']
calc_vif(df_new[numerical_features_fe1])
```

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
  After removing the Total Charges column, we can observe that the level of multicollinearity has been significantly reduced.</p>
</div>


```python
df_num = df_new[numerical_features]

# Feature Scaling (mandatory before PCA)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_num)

# applying PCA
pca = PCA()
pca.fit(scaled_data)

# Variance ratio
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# Plot scree plot
plt.plot(range(1, len(explained_variance)+1), cumulative_variance, marker='o')
plt.xticks(range(1, len(explained_variance)+1)) 
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Choosing PCA Components")
plt.grid(True)
plt.show()
```

```python
n = 2
cum_var_n = cumulative_variance[n-1]
print(f"Cumulative Variance at Component {n}: {cum_var_n:.2f}")
```

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
  From the PCA results, we can see that with only two dimensions, it is already possible to represent almost all the information from the numerical data, accounting for approximately 98% of the variance.</p>
</div>


```python
pca = PCA(n_components=2)  # try reducing to 2 dimensions
pca_data = pca.fit_transform(scaled_data)

# Convert PCA results into a new DataFrame
pca_df = pd.DataFrame(
    pca_data,
    columns=['PC1', 'PC2']
)

pca_df.head(2)
```

```python
calc_vif(pca_df)
```

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
I would recommend applying Principal Component Analysis (PCA) rather than removing the TotalCharges feature. The reason is that PCA allows us to retain the information from all three features (e.g., tenure, MonthlyCharges, and TotalCharges) by transforming them into new principal components. While the interpretability of these components is admittedly less intuitive compared to simply discarding a variable, PCA is statistically more robust as it mitigates multicollinearity without sacrificing the underlying information contained in the original variables.</p>
</div>


```python
def cramers_v(series1, series2):
    contingency = pd.crosstab(series1, series2)
    chi2 = chi2_contingency(contingency)[0]
    n = contingency_table.sum().sum()
    min_dim = min(contingency_table.shape) - 1
    cramers_v = np.sqrt(chi2 / (n * min_dim))
    return cramers_v

def categorical_corr(df, categorical_features, target_col):
    results = []
    for f in categorical_features:
        cv = cramers_v(df[f], df[target_col])
        results.append({'Features': f, 'Cramers_V': cv})
    return (pd.DataFrame(results)
            .sort_values('Cramers_V', ascending=False)
            .reset_index(drop=True)
           )

cat_features_corr=categorical_corr(df_new,categorical_features, target)
cat_features_corr
```

```python
# Pipeline for numerical features without PCA
numerical_pipeline = Pipeline([
    ('imputer', IterativeImputer(random_state=0,min_value=0)),   # handle missing values
    ('scaler', MinMaxScaler()),                      # scale numerical features
])

# Pipeline for categorical features
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  # handle missing values
    ('encoder', OneHotEncoder(handle_unknown='ignore'))    # encode categorical features
])

# Pipeline for already encoded categorical features
categorical_encoded_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
])

# Combine all pipelines into a ColumnTransformer
preprocessor_1 = ColumnTransformer([
    ('num', numerical_pipeline, num_features),                 # numeric features
    ('cat', categorical_pipeline, cat_features),               # categorical features
    ('cat_enc', categorical_encoded_pipeline, cat_enc_features) # already encoded categorical features
], verbose=True)


# Apply the preprocessing pipeline
X_train_processed_1 = preprocessor_1.fit_transform(X_train)  # fit + transform on training data
X_test_processed_1 = preprocessor_1.transform(X_test)        # only transform on test data
```

```python
def xgboost_classifier(X_train, X_test, y_train, y_test):
    # Configure EarlyStopping
    early_stopping = xgb.callback.EarlyStopping(
        rounds=5,             # Number of rounds without improvement before stopping
        metric_name='logloss',# Metric monitored for early stopping (logloss is common for classification)
        save_best=True,       # Save the best model during training
    )
   

    # Initialize XGBoost model for classification
    xg_clf = xgb.XGBClassifier(
        objective='binary:logistic',  # Binary classification objective
        learning_rate=0.3, # default (0.3)
        n_estimators=300,
        random_state=42,
        eval_metric="logloss",      # Evaluation metric for classification
        callbacks=[early_stopping]
    )

    # Evaluation datasets
    evals = [(X_train, y_train),
             (X_test, y_test)]

    # Train the model with early stopping
    evals_result = {}
    xg_clf.fit(
        X_train, y_train,
        eval_set=evals,
        verbose=True
    )

    return xg_clf, xg_clf.evals_result()
```

```python
def plot_learning_curve(model, evals_result, metric='logloss', model_name='XGBoost'):
    """
    Plot the XGBoost learning curve using training and validation metrics.

    Parameters:
    - model: Trained XGBoost model (XGBClassifier or XGBRegressor).
    - evals_result: Dictionary containing evaluation results from training.
    - metric: Metric to plot (e.g., 'logloss', 'error', 'auc').
    - model_name: Name of the model (default 'XGBoost').
    """

    # Extract training and validation metrics
    train_metric = evals_result['validation_0'][metric]
    valid_metric = evals_result['validation_1'][metric]

    # Number of iterations
    epochs = len(train_metric)
    x_axis = range(epochs)

    # Get the best iteration from the model
    best_iter = model.best_iteration
    best_val_score = valid_metric[best_iter]
    print(f"Best iteration: {best_iter}, Validation {metric}: {best_val_score:.4f}")

    # Plot the learning curve
    plt.figure(figsize=(10, 6))
    plt.plot(x_axis, train_metric, label=f"Training {metric}")
    plt.plot(x_axis, valid_metric, label=f"Validation {metric}")

    # Highlight the best iteration
    plt.axvline(best_iter, color="gray", linestyle="--", label="Optimal Iteration")
    plt.scatter(best_iter, best_val_score, color="red", zorder=5)

    # Tambahkan nilai skor di atas titik merah
    plt.text(best_iter, best_val_score,
             f"{best_val_score:.4f}",
             ha='left', va='bottom', fontsize=12, color='red')

    # Custom x-axis ticks (to make the plot cleaner)
    plt.xticks(range(0, epochs + 1, max(1, epochs // 10)))

    plt.xlabel("Iterations")
    plt.ylabel(f"{metric}")
    plt.title(f"{model_name} Learning Curve")
    plt.legend()
    plt.show()
```

```python
def evaluate_model(model_name, y_train, y_train_pred, y_test, y_test_pred):
    """
    Evaluate a model on both training and test sets and return key performance metrics, 
    including an overall weighted metric (Recall > F1 & Precision > Accuracy).

    Parameters
    ----------
    model_name : Name of the model to be used as the row index in the result DataFrame.
    y_train : True labels for the training set.
    y_train_pred : Predicted labels for the training set.
    y_test : True labels for the test set.
    y_test_pred : Predicted labels for the test set.
    
    Returns
    -------
    pandas.DataFrame
        A DataFrame containing Precision, Recall, F1-Score, and Accuracy 
        for both training and test sets, indexed by the model name.
    """
    # Calculate metrics
    precision_train = precision_score(y_train, y_train_pred, zero_division=0)
    recall_train    = recall_score(y_train, y_train_pred, zero_division=0)
    f1_train        = f1_score(y_train, y_train_pred, zero_division=0)
    acc_train       = accuracy_score(y_train, y_train_pred)

    precision_test  = precision_score(y_test, y_test_pred, zero_division=0)
    recall_test     = recall_score(y_test, y_test_pred, zero_division=0)
    f1_test         = f1_score(y_test, y_test_pred, zero_division=0)
    acc_test        = accuracy_score(y_test, y_test_pred)

    # Define weights: Recall > F1 > Precision > Accuracy
    weights = {"Recall": 0.35, "F1": 0.25, "Precision": 0.25, "Accuracy": 0.15}
    overall_test = (
            weights["Recall"] * recall_test +
            weights["F1"] * f1_test +
            weights["Precision"] * precision_test +
            weights["Accuracy"] * acc_test
    )
    # Collect results
    metrics = {
        "Precision (Train)": precision_train,
        "Recall (Train)": recall_train,
        "F1-Score (Train)": f1_train,
        "Accuracy (Train)": acc_train,
        "Precision (Test)": precision_test,
        "Recall (Test)": recall_test,
        "F1-Score (Test)": f1_test,
        "Accuracy (Test)": acc_test,
        "Overall (Test)": overall_test,
    }

    # Convert dictionary to DataFrame and transpose for readability
    df = pd.DataFrame(metrics, index=[model_name]).T
    return df
```

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
  From the function above, we introduce a new weighted metric to evaluate the overall performance of our model. In this weighting scheme, we assign the highest priority to recall, followed by F1-score, precision, and finally accuracy.

```python

def plot_confusion_matrix(models_preds, y_true, titles):
    """
    Visualize the confusion matrix for one or multiple models side by side..

    Parameters
    ----------
    models_preds : list of array-like
        Predicted labels for each model.
    y_true : array-like
        True labels.
    titles : list of str
        Titles for each confusion matrix corresponding to the models.
    """
    n_models = len(models_preds)
    n_cols = 2  # Number of columns in the plot grid
    n_rows = math.ceil(n_models / n_cols)  # Number of rows needed

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(6 * n_cols, 5 * n_rows))
    axes = axes.ravel()  # Flatten axes array for easy indexing

    for i, (y_pred, title) in enumerate(zip(models_preds, titles)):
        cm = confusion_matrix(y_true, y_pred)
        
        # Custom labels for binary confusion matrix
        labels = np.array([
            [f"TN\n{cm[0,0]}", f"FP\n{cm[0,1]}"],
            [f"FN\n{cm[1,0]}", f"TP\n{cm[1,1]}"]
        ])
        
        sns.heatmap(cm, annot=labels, fmt="", cmap="Blues", cbar=False, ax=axes[i])
        axes[i].set_xlabel("Predicted Label")
        axes[i].set_ylabel("True Label")
        axes[i].set_title(f"Confusion Matrix - {title}")

    # Remove empty axes if number of models is odd
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.subplots_adjust(hspace=0.4, wspace=0.3)
    plt.show()

# List of model predictions
models_preds = [y_test_pred_1, y_test_pred_2, y_test_pred_3, y_test_pred_4]
# Titles corresponding to each model
titles = ["XGBoost FE 1", "XGBoost FE 2", "XGBoost FE 3", "XGBoost FE 4"]

# Plot confusion matrices for all models
plot_confusion_matrix(models_preds, y_test, titles)
```

```python
def plot_precision_recall_vs_threshold(precision, recall, thresholds, title='Precision and Recall Trade-off Across Thresholds'):
    """
    Plot Precision and Recall versus Confidence Threshold using Plotly.

    Parameters
    ----------
    precision : array-like
        Precision values for each threshold (from sklearn.metrics.precision_recall_curve)
    recall : array-like
        Recall values for each threshold
    thresholds : array-like
        Confidence thresholds
    title : str, optional
        Plot title (default is 'Precision and Recall Trade-off Across Thresholds')
    """
    fig = go.Figure()

    # Plot Precision
    fig.add_trace(go.Scatter(
        x=thresholds,
        y=precision[:-1],  # skip last to match thresholds length
        mode='lines',
        name='Precision',
        line=dict(color='blue'),
        hovertemplate='Threshold: %{x:.2f}<br>Precision: %{y:.2f}<extra></extra>'
    ))

    # Plot Recall
    fig.add_trace(go.Scatter(
        x=thresholds,
        y=recall[:-1],
        mode='lines',
        name='Recall',
        line=dict(color='red'),
        hovertemplate='Threshold: %{x:.2f}<br>Recall: %{y:.2f}<extra></extra>'
    ))

    # Layout
    fig.update_layout(
        title=title,
        xaxis_title='Confidence Threshold',
        yaxis_title='Value',
        legend_title='Metrics',
        template='plotly_white'
    )

    fig.show(renderer='iframe')
```

In the previous model, we applied OneHotEncoder to transform categorical data. However, this approach resulted in a high-dimensional dataset, which slowed down the training process and and increased memory usage.
<br>
<br>
Tree-based models are generally insensitive to the numerical ordering of categorical variables. This means we can safely use Ordinal Encoding without introducing bias, as these models can effectively split on ordinal features regardless of their assigned numeric values. Using Ordinal Encoding not only preserves model performance but also reduces dimensionality, leading to faster training compared to One-Hot Encoding.
<br>
<br>
Now, we will train an XGBoost model using data encoded with an OrdinalEncoder. The categorical values are ordered using a custom scheme in which ‚ÄúNo‚Äù is assigned a lower rank than ‚ÄúYes‚Äù. This approach improves consistency in interpretation and enables clearer explanations when applying Explainable AI methods. After encoding, we will compare the results to assess whether there are significant differences in the evaluation metrics.
  </p>
</div>


```python
# Define mapping tuples for different types of categorical features
yes_no_map = ('No', 'Yes')
internet_service_map = ('No', 'DSL', 'Fiber optic')
add_service_map = ('No internet service', 'No', 'Yes')
contract_map = ('Month-to-month', 'One year', 'Two year')
payment_method_map = ('Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic)')

# Group features according to which mapping they use
feature_groups = {
    yes_no_map: ['PaperlessBilling', 'Dependents', 'Partner'],
    add_service_map: ['OnlineSecurity', 'TechSupport', 'OnlineBackup', 'DeviceProtection', 'StreamingMovies', 'StreamingTV'],
    internet_service_map: ['InternetService'],
    contract_map: ['Contract'],
    payment_method_map: ['PaymentMethod']
}

# List of categorical features in the desired order
cat_features = [
    'Contract', 'OnlineSecurity', 'TechSupport', 'InternetService', 'PaymentMethod',
    'OnlineBackup', 'DeviceProtection', 'StreamingMovies', 'StreamingTV',
    'PaperlessBilling', 'Dependents', 'Partner'
]

# Generate cat_orders: for each feature, find the corresponding mapping tuple
cat_orders = [k for f in cat_features 
              for k, v in feature_groups.items() 
              if f in v]

cat_orders
```

```python
# Pipeline for numerical features without PCA
numerical_pipeline = Pipeline([
    ('imputer', IterativeImputer(random_state=0,min_value=0)),   # handle missing values
    ('scaler', MinMaxScaler()),                      # scale numerical features
])

# Pipeline for categorical features
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  # handle missing values
    ('encoder', OrdinalEncoder(categories=cat_orders,
                               handle_unknown='use_encoded_value',
                               unknown_value=-1))    # encode categorical features
])

# Pipeline for already encoded categorical features
categorical_encoded_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
])

# Combine all pipelines into a ColumnTransformer
preprocessor_3 = ColumnTransformer([
    ('num', numerical_pipeline, num_features),                 # numeric features
    ('cat', categorical_pipeline, cat_features),               # categorical features
    ('cat_enc', categorical_encoded_pipeline, cat_enc_features) # already encoded categorical features
], verbose=True)


# Apply the preprocessing pipeline
X_train_processed_3 = preprocessor_3.fit_transform(X_train)  # fit + transform on training data
X_test_processed_3 = preprocessor_3.transform(X_test)        # only transform on test data
```

```python
# Pipeline for numerical features with PCA
numerical_pipeline_pca = Pipeline([
    ('imputer', IterativeImputer(random_state=0, min_value=0)),   # handle missing values
    ('scaler', MinMaxScaler()),                      # scale numerical features
    ('pca', PCA(n_components=2, random_state=0))     # reduce to 2 components
])


# Pipeline for categorical features
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  # handle missing values
    ('encoder', OrdinalEncoder(categories=cat_orders,
                               handle_unknown='use_encoded_value',
                               unknown_value=-1))    # encode categorical features
])

# Pipeline for already encoded categorical features
categorical_encoded_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
])


# Combine all pipelines into a ColumnTransformer
preprocessor_4 = ColumnTransformer([
    ('num', numerical_pipeline_pca, num_features),                 # numeric features (PCA)
    ('cat', categorical_pipeline, cat_features),               # categorical features
    ('cat_enc', categorical_encoded_pipeline, cat_enc_features) # already encoded categorical features
], verbose=True)


# Apply the preprocessing pipeline
X_train_processed_4 = preprocessor_4.fit_transform(X_train)  # fit + transform on training data
X_test_processed_4 = preprocessor_4.transform(X_test)        # only transform on test data
```

```python
# Get feature names from each pipeline
num_cols = [f'PC{i+1}' for i in range(2)]    # numerical features after PCA
cat_cols = preprocessor_4.named_transformers_['cat']['encoder'].get_feature_names_out(cat_features)  # one-hot encoded categorical features
cat_enc_cols = cat_enc_features  # already encoded categorical features

# Combine all feature names
all_feature_names_4 = list(num_cols) + list(cat_cols) + list(cat_enc_cols)

# Create DataFrame from the transformed arrays
X_train_df_4 = pd.DataFrame(X_train_processed_4, columns=all_feature_names_4)
X_test_df_4 = pd.DataFrame(X_test_processed_4, columns=all_feature_names_4)

print('Shape:',X_train_df_4.shape)
X_train_df_4.head()
```

## 10.2 XGBoost - Hyperparameter Tuning 

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
In this hyperparameter tuning, we use Grid Search for selecting the best model parameters based on the recall score. The reason for choosing recall is that it prioritizes the model‚Äôs ability to capture positive cases.
  </p>
</div>


```python
# Define scoring dictionary
scoring = {
    'f1': make_scorer(f1_score, pos_label=1), #default average= 'binary'
    'recall': make_scorer(recall_score, pos_label=1)
}

scale_pos_weight = sum(y_train == 0) / sum(y_train == 1)
scale_pos_weight

# Base model
xgb_clf = xgb.XGBClassifier(
    objective='binary:logistic',
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=scale_pos_weight,
)

param_grid = {
        'n_estimators':range(8,15),
        'max_depth': range(3,8),
        'learning_rate': [0.1, 0.2, 0.3],
        'subsample': [0.8, 1.0],
        'colsample_bytree': [0.8, 1.0],
        'min_child_weight': [1, 3, 5],
    }

cv_strategy = StratifiedKFold(n_splits=5, shuffle=False)

grid_search = GridSearchCV(
    estimator=xgb_clf,
    param_grid=param_grid,
    scoring=scoring,      # multiple metrics
    refit='recall',           # model will be selected based on recall
    return_train_score=True,
    verbose=0,
    cv=cv_strategy,
    n_jobs=-1
)

# Fit the grid search using the preprocessed training data (ordinal encoder without PCA)
grid_search.fit(X_train_processed_3, y_train)
```

<div style="background-color:#c9d5ff; padding: 20px;">
  <p style="font-size:20px; font-family:verdana; line-height: 1.7em">
we use <strong>scale_pos_weight</strong> to handle data imbalance. After addressing the imbalance, we can still adjust the decision threshold to further improve the model‚Äôs performance.
  </p>
</div>

```python
np.random.seed(42) 
search_spaces = {
        "n_estimators": Integer(8, 11),
        "max_depth": Integer(3, 8),
        "learning_rate": Categorical([0.1, 0.2, 0.3]),
        "subsample": Categorical([0.8, 1.0]),
        "colsample_bytree": Categorical(list(np.arange(0.8, 1.01, 0.1))),
        "min_child_weight": Integer(1, 5)
    }
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

opt = BayesSearchCV(
        estimator=xgb_clf,
        search_spaces=search_spaces,
        scoring="recall",  # optimize primarily for recall
        cv=cv_strategy,
        n_jobs=-1,
        n_iter=20,
        random_state=42,
        refit=True,        # retrain best model after search
        return_train_score=True,
        verbose=0
    )

opt.fit(X_train_processed_3, y_train)
```
